```python
class Car:
    def __init__(self):
        self.color = 'Black'
        self.wheel = 4
        self.spare_wheel = 1
        self.airbags_support = True
        self.fuel_type = 'Petrol'
        self.car_type = 'Sedan'
        self.gear_type = 'Automatic'
        print('Your car is ready!!')

    def auto_park(self):
        from IPython.display import Image, display
        url = 'https://media.tenor.com/OK4IzfGb50IAAAAM/parkir-anak-kecil.gif'
        display(Image(url=url))
        
    def engine_on(self):
        import pygame
        pygame.init()
        path = r'''C:\Users\Aksha\Downloads\car-sound-effect-126709.mp3'''
        music = pygame.mixer.Sound(path)
        music.play()
         
    def speed_up(self):
        c = 10
        while c<=200:
            import time
            print(f'Speed is Increasing by {c} km/hr')
            c+= 10
            display(clear= True)
            time.sleep(0.5)

            if c >=100:
                print('You are crossing speed limit')
                break
               
    def drift_barke(self,drift=True):
        if drift == True:
            import pygame
            pygame.init()
            path = r'''C:\Users\Aksha\Downloads\-110063.mp3'''
            music = pygame.mixer.Sound(path)
            music.play()
            import IPython
            url = '''https://media.tenor.com/puPGAEN97U0AAAAM/lol-mekro.gif'''
            display(IPython.display.Image(url))
        else:
            print('Brake Applied!!')

         
        
```

```python
class ATM:
    def __init__(self):
        print('Object created successfully!!')
        self.pin = 1111     # public variable
        self.balance = 2000

    def change_pin(self):
        self.pin = input("Enter new PIN: ")
        print('Your PIN has been changed.')

    def show_balance(self):
        print(f"Your current balance is Rs {self.balance}")

    def atm_performance_stats(self):
        print('ATM working fine 100%')

        
```

```python
icici = ATM()
```

```python
icici.show_balance()
```

```python
icici.change_pin()
```

```python
icici.atm_performance_stats()
```

```python
class ATM_version_1:
    def __init__(self):
        print('Object created successfully!!')
        self._pin = 1111       #protected variable uses "_"
        self.__balance = 2000   #private variable uses "__"

    def change_pin(self):
        self.pin = input("Enter new PIN: ")
        print('Your PIN has been changed.')

    def show_balance(self):
        print(f"Your current balance is Rs {self._balance}")

    def atm_performance_stats(self):
        print('ATM working fine 100%')
```
